```python
from pylab import *
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('grb_data.csv')
data = data.set_index('GRB')
```

```python
data.count()
```

```python
data.mean()
```

```python
data.median()
```

```python
data.std()
```

```python
data.var()
```

```python
histo = data['RA'].plot.hist(bins=50)
title('Right Ascension')
xlabel('Minutes')
fig = histo.get_figure()
fig.savefig('RA.pdf')
```

```python
histo = data['Dec'].plot.hist(bins=50)
title('Declination')
xlabel('Degrees')
fig = histo.get_figure()
fig.savefig('Dec.pdf')
```

```python
histo = data['T90'].plot.hist(bins=50)
title('T90')
xlabel('Seconds')
fig = histo.get_figure()
fig.savefig('T90.pdf')
```

```python
histo = data['Fluence'].plot.hist(bins=50)
title('Fluence')
xlabel('10**-7 erg/cm2')
fig = histo.get_figure()
fig.savefig('Fluence.pdf')
```

```python
histo = data['PhotonIndex'].plot.hist(bins=50)
title('Photon Index')
fig = histo.get_figure()
fig.savefig('PhotonIndex.pdf')
```

```python
histo = data['Redshift'].plot.hist(bins=50)
title('Redshift')
xlabel('z')
fig = histo.get_figure()
fig.savefig('Redshift.pdf')
```

```python
for column in data:
    if np.issubdtype(data[column].dtype, np.number):
        histo = data[column].plot.hist(bins=50)
```
